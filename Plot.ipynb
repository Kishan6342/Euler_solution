```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Output20.csv')

# Plot error vs t
plt.figure(figsize=(8, 5))
plt.plot(df['t'], df['error'], marker='o')
plt.xlabel('t')
plt.ylabel('Error')
plt.title('Euler Method Error vs Analytical Solution')
plt.grid(True)
plt.tight_layout()
plt.show()
```
